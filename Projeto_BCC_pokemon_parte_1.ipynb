```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

poke = pd.read_csv('http://professor.ufabc.edu.br/~ronaldo.prati/bases/Pokemon.csv', sep=',' )
```

```
#Remoção da coluna "#"
poke=poke.drop(['#'],axis=1)
poke
```

```
#Número de pokemons por geração
geracoes = poke.groupby("Generation").count()
print (geracoes, "No total de",  geracoes.sum(),"pokémons")
poke.columns
```

```
#Quantidade de pokémons para cada geração utilizando gráfico de barras da biblioteca seaborn
plt.figure(figsize=(10,6))
sns.countplot(poke['Generation'])
plt.show()
```

```
#Histograma com a quantidade de pokémons pra cada geração, para comparar a visualização com o gráfico de barras acima
plt.figure(figsize=(9,5))
plt.xlabel('Gerações')
plt.ylabel('Quantidade máxima de pokémons')
plt.title('Quantidade de pokémon por geração')
plt.hist(poke['Generation'], bins=6)
plt.show()
```

```
#Histograma com os pokémons do tipo 1. Um fato interessante de notar no histograma é a baixa incidência de
#pokémons voadores, sendo que no pokémon, a maioria é exatamente deste tipo, porém, eles estão localizados no Tipo 2.
plt.figure(figsize=(14,4))
plt.ylabel('Total de pokémons por tipo')
plt.title('Pokémons do tipo 1')
plt.hist(poke['Type 1'], bins=18)
plt.show()
```

```
#Histograma com os pokémons do tipo 2. Neste histograma, a maioria tem o tipo voador porque a maioria dos pokémons
#tem o tipo "voador" como secundário, depois do elementar, por exemplo o Dragonite, é um dragão do tipo voador.
plt.figure(figsize=(14,4))
plt.ylabel('Total de pokémons por tipo')
plt.title('Pokémons do tipo 2')
plt.hist(poke['Type 2'], bins=18)
plt.show()
```

```
#Boxplot mostrando como os status dos pokémons se concentram de maneira geral, com alguns extrapolando a média
retirar = poke.drop(['Generation', 'Total','Type 1','Type 2','Legendary' ], axis=1)
plt.title('Distribuição dos atributos dos pokémons')
sns.boxplot(data=retirar)
plt.show()
```

```
#Este gráfico de distribuição mostra como se distribui a soma dos atributos dos pokémons(ataque, defesa, vida etc.)
sns.distplot(poke['Total'],axlabel='Distribuição da soma de todos os atributos')
```

```
#Faz um graficoo de barras com a média por geração do atributo digitado
atributo = input("Digite o nome do atributo para ver a média, ex: Speed")
media1 = poke.loc[poke['Generation'] == 1][atributo].mean()
media2 = poke.loc[poke['Generation'] == 2][atributo].mean()
media3 = poke.loc[poke['Generation'] == 3][atributo].mean()
media4 = poke.loc[poke['Generation'] == 4][atributo].mean()
media5 = poke.loc[poke['Generation'] == 5][atributo].mean()
media6 = poke.loc[poke['Generation'] == 6][atributo].mean()

eixoY = [media1,media2,media3,media4,media5,media6]
eixoX = ['Gen1','Gen2','Gen3','Gen4','Gen5','Gen6']

fig = plt.figure(figsize = (10, 5))
plt.bar(eixoX, eixoY, color ='green',
        width = 0.5)

plt.xlabel("Geração")
plt.ylabel("Média do atributo")
plt.title("Média dos atributos por geração")
plt.show()
```
